```python
data['StartYear'] = data['Start Date'].dt.year

df16 = data[data['StartYear'] == 2016]
df16.head()
```

```python
df16one = df16[df16['Station Name'] == 'PALO ALTO CA / WEBSTER #2']
df16one = df16one.reset_index()
df16one = df16one.rename(columns={'Start Date': 'Start', 'End Date': 'End', 'index': "ID"})
df16one.head()
```

## Let's go!

```python
from collections import OrderedDict

df16one['End_d']=pd.DatetimeIndex(df16one['End']).day
df16one['Start_d']=pd.DatetimeIndex(df16one['Start']).day
```

```python
df16one.head(1)
```

```python
df16one.dtypes
```

```python
df1 = df16one[df16one.End_d==df16one.Start_d].loc[:,['index', 'Start','End']]  #the obs. of which the duration < 1day
df2 = df16one[df16one.End_d!=df16one.Start_d]
```

```python
for i in df2.index:   #Expand it in to multiple rows.
    days=df2.loc[i,:].End_d-df2.loc[i,:].Start_d+1
    start_d_str=df2.loc[i,:].Start.strftime('%Y-%m-%d')

    temp_df=pd.DataFrame(OrderedDict({'OperID': df2.loc[i,:].ID,
              'Start': pd.date_range('%s 00:00:00'%start_d_str, periods=days),
              'End':   pd.date_range('%s 23:59:59'%start_d_str, periods=days)}))

    temp_df.loc[0,'Start'] = df2.loc[i,'Start']
    temp_df.loc[days-1, 'End'] = df2.loc[i,'End']
    df1=df1.append(temp_df)
```
